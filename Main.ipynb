```python
for i in range(Number):
    AddExpense = input().split(",")
    if os.path.exists(File):
        df = pd.read_csv(File)
        AddExpense[1] = int(AddExpense[1])
        df.loc[len(df)] = AddExpense
        df.to_csv(File,index=False)
    else:
        df= pd.DataFrame({"Expense":[AddExpense[0]],"Amount":[int(AddExpense[1])],"Category":[AddExpense[2]],"Date":[AddExpense[3]]})
        df.to_csv(File,index=False)
df["Date"] = pd.to_datetime(df["Date"],format="%Y-%m-%d", errors="coerce")
df.to_csv(File,index=False)
```
